```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows',10)

grupos_mundial=pd.read_csv('Fifa_Worldcup_2022_Groups.csv')

grupos_mundial['Puntos']=0

pts = np.random.randint(0,10,size=32)

probabilidad_de_ganar = np.random.uniform(0,1,size=32)

grupos_mundial['Puntos'] = pts

grupos_mundial['Probabilidad']=probabilidad_de_ganar

print(grupos_mundial.columns)

print(grupos_mundial[['Country_Name_Short','Puntos']])

print(grupos_mundial['Probabilidad'].std())

print(grupos_mundial['Probabilidad']*grupos_mundial['Puntos'])
```

```python
grupos_mundial.sort_values('Group',ascending=True,inplace=True)
gf = np.array([1,4,5,5,9,4,2,1,5,3,2,2,6,3,1,1,9,3,6,4,1,2,4,4,3,5,4,4,6,5,2,4])
pts = np.array([0,4,6,7,7,3,5,1,6,3,4,4,6,6,1,4,4,3,4,6,4,0,7,5,6,1,6,4,6,3,4,4])
grupos_mundial['Goles_Hechos']=gf
grupos_mundial['Puntos']=pts
grupos_mundial
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt 
x=grupos_mundial['Goles_Hechos']
y=grupos_mundial['Puntos']
grupos_mundial.plot(kind='scatter',x='Goles a favor',y='Puntos')
```

```python
x = sm.add_constant(x)
modelo=sm.OLS(y,x).fit()
modelo.predict(x).round(0)
modelo.summary()
```

```python
y_pred = 0.459 * x['Goles_Hechos']+2.4664
```

```python
import seaborn as sns
#Graficar los puntos
plt.figure(figsize=(6,4),tight_layout=True)
sns.scatterplot(x=x['Goles_Hechos'],y=y)
#graficar la linea
sns.lineplot(x=x['Goles_Hechos'],y=y_pred,color='red')
#Ejes
plt.xlim(0)
plt.ylim(0)
plt.savefig('linear_regresion')
plt.show 
```
